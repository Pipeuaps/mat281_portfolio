5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
#ordenamiento de datos
etiquetas=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
df_reduce=df[etiquetas]
df_reduce
```

```python
#suma de las ventas totales en el año de cada producto
valores=[]
for etiqueta in etiquetas:
    valores.append(df_reduce[etiqueta].sum())
valores
```

```python
#grafico circular
plt.pie(valores, labels=etiquetas,
        autopct='%1.1f%%', shadow=True, startangle=90)
```
